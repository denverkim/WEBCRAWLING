```python
titanic = pd.read_csv('https://raw.githubusercontent.com/sehakflower/data/main/titanic.csv', delimiter='\t')
titanic.head()
```

```python
# 전체 컬럼이름 변경 ('passengerId', 'survived', 'pclass', 'name', 'sex', 'age', 'sib', 'parch', 'ticket', 'fare', 'cabin', 'embarked’)
titanic.columns = ['passengerId', 'survived', 'pclass', 'name', 'sex', 'age', 'sib', 'parch', 'ticket', 'fare', 'cabin', 'embarked']
titanic.head()
```

```python
# age 컬럼에서 각 행의 널값 여부 (isna or isnull), 널값인 행 출력
print(titanic['age'].isna())
print(titanic[titanic['age'].isna()])
```

```python
# age 컬럼에서 널값의 개수는?
titanic['age'].isna().sum()
```

```python
# age컬럼이 널값인 사람의 age, name은?
titanic[titanic['age'].isna()][['age', 'name']]
```

```python
# age 컬럼에서 각 행의 널값 여부 (notnull), 널값이 아닌 행 출력
print(titanic['age'].notnull())
print(titanic[titanic['age'].notnull()])
```

```python
# age컬럼이 널값이 아닌 사람의 age, name은?
titanic[titanic['age'].notnull()][['age', 'name']]
```

```python
# cabin컬럼의 널값을 N으로 채우기
titanic['cabin'].fillna('N')
```

```python
# cabin컬럼의 첫번째 글자만 출력
titanic['cabin'].str[0]
```

```python
# age컬럼의 널값을 age평균으로 채우기

titanic.age.fillna(titanic.age.mean(), inplace=True)
titanic.age
```
