# 세금계산 프로그램
기본공제 10,000불
부양가족당 3,000불 공제
세금 20%

```python
print("세금 계산 프로그램")
print("="*20)
income = input("\n수입은 얼마인가요? ")
num = input("부양가족은 몇명인가요? ")
tax = (int(income) - 10000 - 3000 * int(num))*.2
print("내셔야 할 세금은 ", tax, "입니다")
```
